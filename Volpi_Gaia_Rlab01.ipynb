5. Plot the monthly average trip duration

```r
by_month = data1 %>% group_by(month=month(data1$starttime, label=TRUE)) %>% summarise(duration=mean(tripduration),std=sd(tripduration))

ggplot(by_month, aes(x=month, y=duration)) +
    geom_point() + 
    geom_errorbar(aes(ymin=duration-std, ymax=duration+std), width = 0.1) 
```
